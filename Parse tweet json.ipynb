### Mode selection
- Mode 1: Combine and parse all JSON files into a single CSV output file
- Mode 2: Parse JSON files individually and generate separate CSV output files

```python
all_files = glob.glob(os.path.join(input_dir, "*.json"))

for file in all_files: #Load each of the JSON files from the input folder
    filename = os.path.basename(file)
    
    with open(file) as f:
    
        json_data = twarc.expansions.flatten(json.loads(f.read()))
        df = pd.json_normalize(json_data)
        df = extract_referenced_tweets(df)
        df = add_full_text(df)
        df = extract_expanded_urls(df)
        df = extract_media_urls(df)
        print("The length of the CSV file", filename, "after parsing: " + str(len(df)))
        #Write the merged CSV file after parsing to the output folder
    df.to_csv(output_dir + filename)  
```

```python
all_files = glob.glob(os.path.join(input_dir, "*.json"))
parsed_df = []
for file in all_files: #Load each of the JSON files from the input folder
    #print(file)
    with open(file) as f:
        json_data = twarc.expansions.flatten(json.loads(f.read()))
        df = pd.json_normalize(json_data)
        df = extract_referenced_tweets(df)
        df = add_full_text(df)
        df = extract_expanded_urls(df)
        df = extract_media_urls(df)
        parsed_df.append(df)

parsed_data = pd.concat(parsed_df, ignore_index=True) #concat all the JSON files in the list into one dataframe
print("Total length of the merged CSV file after parsing: " + str(len(parsed_data)))
#Write the merged CSV file after parsing to the output folder
parsed_data.to_csv(output_dir + output_name)
```
